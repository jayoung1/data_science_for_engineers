```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import metrics
from sklearn import datasets
```

```python
digits = datasets.load_digits(as_frame = True)
```

```python
digits
```

```python
# flatten the images
n_samples = len(digits.images)
df = digits.images.reshape((n_samples, -1))
```

```python
df.shape
```

```python
x = df
y = digits["target"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

print(rf.score(x_train, y_train))

y_pred_rf = rf.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred_rf))
```

```python
print(metrics.confusion_matrix(y_pred_rf, y_test))
print('-----')
print(metrics.classification_report(y_pred_rf, y_test))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier()

mlp.fit(x_train, y_train)

print(mlp.score(x_train, y_train))

y_pred_mlp = mlp.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred_mlp))
```

```python
print(metrics.confusion_matrix(y_pred_mlp, y_test))
print('-----')
print(metrics.classification_report(y_pred_mlp, y_test))
```

```python
mlp = MLPClassifier(verbose=True)

mlp.fit(x_train, y_train)

print(mlp.score(x_train, y_train))

y_pred_mlp = mlp.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred_mlp))
```

```python
from sklearn.model_selection import GridSearchCV

grid_params = {
    "activation":["relu","logistic"],
    "hidden_layer_sizes":[(100,),(200,),(100,100,)],
    "alpha":[0.0001,0.005],
    "learning_rate":["constant","adaptive"],
    "max_iter":[1000]
}

gs = GridSearchCV(
    MLPClassifier(),
    grid_params,
    cv = 2
)

gs_results = gs.fit(x_train, y_train)

print(gs_results.best_score_)
print(gs_results.best_estimator_)
print(gs_results.best_params_)
```

```python
mlp = MLPClassifier(alpha=0.005, hidden_layer_sizes=(200,), max_iter=1000)

mlp.fit(x_train, y_train)

print(mlp.score(x_train, y_train))

y_pred_mlp = mlp.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred_mlp))
```
